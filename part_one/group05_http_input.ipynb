```python
import socket
import time

# הגדרות יעד (שרת בדיקה אמיתי כדי לייצר תעבורה)
target_host = "httpbin.org"
target_port = 80

print(f"Starting traffic generation to {target_host}...")

# מעבר על כל שורה ב-CSV ושליחת ההודעה
for index, row in df.iterrows():
    # נתמקד רק בהודעות שהלקוח שולח (requests) כדי לייצר תעבורה תקינה
    if "client" in row['src_app']:
        try:
            # יצירת סוקט TCP
            client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
            client.connect((target_host, target_port))
            
            # שליחת ההודעה מה-CSV
            # אנחנו מוסיפים Host header כדי שהשרת יבין את הבקשה
            msg = f"{row['message']} HTTP/1.1\r\nHost: {target_host}\r\n\r\n"
            client.send(msg.encode())
            
            print(f"Sent msg_id {row['msg_id']}: {row['message']}")
            
            # סגירת החיבור
            client.close()
            
            # המתנה קצרה לפי ה-timestamp ב-CSV (אופציונלי)
            time.sleep(0.5) 
            
        except Exception as e:
            print(f"Error sending message {row['msg_id']}: {e}")

print("Traffic generation finished.")
```

```python
import socket
import time

# הגדרת כתובת היעד (שרת בדיקה אמיתי באינטרנט)
target_host = "httpbin.org"
target_port = 80

print(f"Starting traffic generation to {target_host}...")

# לולאה שעוברת על כל השורות בטבלה שלך (df)
for index, row in df.iterrows():
    # אנחנו שולחים רק את ההודעות של ה-Client (הדפדפן) כדי לייצר תעבורה יוצאת
    if "client" in row['src_app']:
        try:
            print(f"Connecting to send msg_id {row['msg_id']}...")
            
            # 1. יצירת חיבור אינטרנט אמיתי (Socket)
            client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
            client.connect((target_host, target_port))
            
            # 2. הכנת ההודעה לשליחה
            # (אנחנו מוסיפים שורות טכניות שהשרת צריך כדי להבין אותנו)
            http_request = f"{row['message']} HTTP/1.1\r\nHost: {target_host}\r\nConnection: close\r\n\r\n"
            
            # 3. שליחה בפועל
            client.send(http_request.encode())
            print(f"--> Sent: {row['message']}")
            
            # 4. סגירת החיבור
            client.close()
            
            # המתנה קצרה בין הודעה להודעה כדי שיהיה ברור ב-Wireshark
            time.sleep(1)
            
        except Exception as e:
            print(f"Error: {e}")

print("Traffic generation finished!")
```

```python
import socket
import time
import pandas as pd

# 1. יצירת הנתונים ידנית (כדי לוודא שהטבלה לא ריקה)
data = {
    'msg_id': [1, 3, 5],
    'src_app': ['client_browser', 'client_browser', 'client_browser'],
    'message': ['GET /index.html', 'GET /images/logo.png', 'POST /login_attempt']
}
df = pd.DataFrame(data)

print("Data loaded successfully!")
print(f"Found {len(df)} messages to send.")

# 2. הגדרות שליחה
target_host = "httpbin.org"
target_port = 80

print(f"Starting traffic generation to {target_host}...")

# 3. לולאת שליחה
for index, row in df.iterrows():
    try:
        print(f"Connecting to send msg_id {row['msg_id']}...")
        
        client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        client.connect((target_host, target_port))
        
        # בניית בקשת HTTP תקינה
        http_request = f"{row['message']} HTTP/1.1\r\nHost: {target_host}\r\nConnection: close\r\n\r\n"
        
        client.send(http_request.encode())
        print(f"--> Sent: {row['message']}")
        
        client.close()
        time.sleep(1) # המתנה כדי שיהיה ברור ב-Wireshark
        
    except Exception as e:
        print(f"Error: {e}")

print("Traffic generation finished!")
```
