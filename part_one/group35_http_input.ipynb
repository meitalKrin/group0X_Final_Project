```python
df = pd.read_csv('group35_http_input .csv')
```

```python
df['tcp_header'] = "SrcPort: 54321, DstPort: 80, Seq: " + df['msg_id'].astype(str)
```

```python
df['ip_header'] = "SrcIP: 192.168.1.10, DstIP: 10.0.0.5, Proto: TCP"
```

```python
print("Data Encapsulation Simulated:")
```

```python
df[['message', 'tcp_header', 'ip_header']].head()
```

```python
import socket
import time


target_host = "httpbin.org"
target_port = 80

print(f"Starting traffic generation to {target_host}...")


for index, row in df.iterrows():
    if "client" in row['src_app']:
        try:
            print(f"Connecting to send msg_id {row['msg_id']}...")
            
            client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
            client.connect((target_host, target_port))
            
            http_request = f"{row['message']} HTTP/1.1\r\nHost: {target_host}\r\nConnection: close\r\n\r\n"
   
            client.send(http_request.encode())
            print(f"--> Sent: {row['message']}")
            
            client.close()
            
            time.sleep(1)
            
        except Exception as e:
            print(f"Error: {e}")

print("Traffic generation finished!")
```
